##### Checking if all files of the same level have the same number of columns

```python
import os
import csv
base = '/home/dev/Desktop/Venus/Data'
directories = ["vcors_2001", "vcors_2002", "vcors_2003", "vcors_2004"]
subdirectories = ["l2"]

for subdir in subdirectories:
    column_counts = set()  # Store unique column counts for each subdirectory

    for directory in directories:
        path = os.path.join(base,directory, subdir)
        if not os.path.exists(path):
            continue

        for file_name in os.listdir(path):
            file_path = os.path.join(path, file_name)

            with open(file_path, 'r', newline='') as file:
                reader = csv.reader(file)
                data = list(reader)
                num_columns = len(data[0]) if data else 0
                column_counts.add(num_columns)

        if len(column_counts) == 1:
            print(f"All l2 files in {directory} have the same number of columns: {column_counts.pop()}")
        else:
            print(f"Not all l2 files in {directory} have the same number of columns.")

print("Done.")
```

```python
import os
import csv
base = '/home/dev/Desktop/Venus/Data'
directories = ["vcors_2001", "vcors_2002", "vcors_2003", "vcors_2004"]
subdirectories = ["l3", "l4"]
subsubdirectories = ["ingress","egress"]

for subdir in subdirectories:
    for subsubdir in subsubdirectories:
        column_counts = set()  # Store unique column counts for each subdirectory

        for directory in directories:
            path = os.path.join(base,directory, subdir,subsubdir)
            if not os.path.exists(path):
                continue

            for file_name in os.listdir(path):
                file_path = os.path.join(path, file_name)

                with open(file_path, 'r', newline='') as file:
                    reader = csv.reader(file)
                    data = list(reader)
                    num_columns = len(data[0]) if data else 0
                    column_counts.add(num_columns)

            if len(column_counts) == 1:
                print(f"All {subdir}_{subsubdir} files in {directory} have the same number of columns: {column_counts.pop()}")
            else:
                print(f"Not all {subdir}_{subsubdir} files in {directory} have the same number of columns.")

print("Done.")
```
